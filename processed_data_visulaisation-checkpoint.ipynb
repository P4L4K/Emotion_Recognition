```python
import joblib
import numpy as np
import matplotlib.pyplot as plt

# Load the preprocessed data files
X_train_par, X_test_par, y_train_par, y_test_par, class_weights_dict_par = joblib.load('preprocessed_data_parallel.joblib')
X_train_seq, X_test_seq, y_train_seq, y_test_seq, class_weights_dict_seq = joblib.load('sequential_preprocessed_data.joblib')

# Display basic information about the data
def print_data_info(X_train, X_test, y_train, y_test, label="Data"):
    print(f"--- {label} ---")
    print(f"X_train shape: {X_train.shape}")
    print(f"X_test shape: {X_test.shape}")
    print(f"y_train shape: {y_train.shape}")
    print(f"y_test shape: {y_test.shape}")
    #print(f"Sample X_train data: {X_train[:2]}")  # Display first two samples
    #print(f"Sample y_train data: {y_train[:2]}")  # Display first two labels

# Print details for parallel model data
print_data_info(X_train_par, X_test_par, y_train_par, y_test_par, label="Parallel Model Data")

# Print details for sequential model data
print_data_info(X_train_seq, X_test_seq, y_train_seq, y_test_seq, label="Sequential Model Data")






```

```python
# Visualize and compare data shapes
def plot_data_shape_comparison():
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))

    axes[0].bar(['X_train', 'X_test', 'y_train', 'y_test'], [X_train_seq.shape[0], X_test_seq.shape[0], y_train_seq.shape[0], y_test_seq.shape[0]], color='blue')
    axes[0].set_title('Sequential Model Data Shapes')
    axes[0].set_ylabel('Number of Samples')
    axes[0].set_xlabel('Data Type')

    axes[1].bar(['X_train', 'X_test', 'y_train', 'y_test'], [X_train_par.shape[0], X_test_par.shape[0], y_train_par.shape[0], y_test_par.shape[0]], color='green')
    axes[1].set_title('Parallel Model Data Shapes')
    axes[1].set_ylabel('Number of Samples')
    axes[1].set_xlabel('Data Type')

    plt.tight_layout()
    plt.show()
# Plot data shape comparison
plot_data_shape_comparison()

# Check the class distributions (if labels are categorical)
def plot_class_distribution(y_train, y_test, label="Data"):
    unique_train_classes, train_class_counts = np.unique(y_train, return_counts=True)
    unique_test_classes, test_class_counts = np.unique(y_test, return_counts=True)

    plt.figure(figsize=(12, 6))

    # Train class distribution
    plt.subplot(1, 2, 1)
    plt.bar(unique_train_classes, train_class_counts, color='blue', alpha=0.7)
    plt.title(f'{label} - Train Class Distribution')
    plt.xlabel('Class')
    plt.ylabel('Count')

    # Test class distribution
    plt.subplot(1, 2, 2)
    plt.bar(unique_test_classes, test_class_counts, color='green', alpha=0.7)
    plt.title(f'{label} - Test Class Distribution')
    plt.xlabel('Class')
    plt.ylabel('Count')

    plt.tight_layout()
    plt.show()

# Plot class distribution for both parallel and sequential models
plot_class_distribution(y_train_seq, y_test_seq, label="Sequential Model")
plot_class_distribution(y_train_par, y_test_par, label="Parallel Model")
```

```python
from tensorflow.keras.layers import Input, Dense, Dropout, LayerNormalization, MultiHeadAttention, GlobalAveragePooling1D, BatchNormalization, Add
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from tensorflow.keras.models import Model
from tensorflow.keras.regularizers import l2
from tensorflow.keras.layers import Embedding
# Step 1: Load the Preprocessed Data
X_train, X_test, y_train_one_hot, y_test_one_hot, class_weights_dict = load('sequential_preprocessed_data.joblib')
emotion_labels = {
    '01': 'neutral', '02': 'calm', '03': 'happy', '04': 'sad',
    '05': 'angry', '06': 'fearful', '07': 'disgust', '08': 'surprised'
}
# Step 2: Define Transformer Encoder with Enhanced Layers
def transformer_encoder(inputs, embed_dim, num_heads, ff_dim, dropout_rate=0.1):
    # Multi-Head Attention Block
    attention_output = MultiHeadAttention(num_heads=num_heads, key_dim=embed_dim)(inputs, inputs)
    attention_output = LayerNormalization(epsilon=1e-6)(attention_output + inputs)  # Residual connection + Layer Norm
    attention_output = Dropout(dropout_rate)(attention_output)

    # Feed Forward Block
    ff_output = Dense(ff_dim, activation='relu', kernel_regularizer=l2(0.0001))(attention_output)
    ff_output = Dropout(dropout_rate)(ff_output)
    ff_output = Dense(embed_dim)(ff_output)
    
    return LayerNormalization(epsilon=1e-6)(attention_output + ff_output)

# Modify input layer to match the data shape
input_layer = Input(shape=(X_train.shape[1], X_train.shape[2]))  # Now expects (180, 180) shaped input

# Apply transformer-based block with more layers
embedding = Dense(128, activation='relu')(input_layer)
transformer_output = transformer_encoder(embedding, embed_dim=128, num_heads=8, ff_dim=256)
transformer_output = transformer_encoder(transformer_output, embed_dim=128, num_heads=8, ff_dim=256)

# Apply Global Average Pooling to the transformer output
gap = GlobalAveragePooling1D()(transformer_output)

# Fully connected layers with adjusted dropout rates and BatchNormalization
dense1 = Dense(256, activation='relu', kernel_regularizer=l2(0.001))(gap)  # Increased units
bn1 = BatchNormalization()(dense1)
dropout1 = Dropout(0.6)(bn1)  # Increased dropout to prevent overfitting
dense2 = Dense(128, activation='relu', kernel_regularizer=l2(0.001))(dropout1)
bn2 = BatchNormalization()(dense2)
dropout2 = Dropout(0.5)(bn2)  # Increased dropout to prevent overfitting

# Output layer
output_layer = Dense(len(emotion_labels), activation='softmax')(dropout2)

# Model compilation with a learning rate scheduler
model = Model(inputs=input_layer, outputs=output_layer)
optimizer = Adam(learning_rate=0.0005)  # Slightly increased learning rate

model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

# Step 3: Set Callbacks with Enhanced Monitor Metrics
early_stopping = EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True)  # Increased patience
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=1e-6)
model_checkpoint = ModelCheckpoint("best_transformer_sequential_model.keras", monitor='val_loss', save_best_only=True)

# Step 4: Train the Model
history = model.fit(
    X_train, y_train_one_hot,
    epochs=100,  # Reduce epochs if overfitting occurs
    batch_size=32,
    validation_data=(X_test, y_test_one_hot),
    class_weight=class_weights_dict,
    callbacks=[early_stopping, reduce_lr, model_checkpoint]
)

# Save the training history and model
import joblib
joblib.dump(history.history, 'history_sequential_enhanced.joblib')

# Step 5: Evaluate the Model
test_loss, test_accuracy = model.evaluate(X_test, y_test_one_hot)
print(f"Test Accuracy: {test_accuracy:.4f}")

# Step 6: Save the Final Trained Model
model.save("final_transformer_sequential_model_enhanced.keras")
```
